```python
# Q5. 요약 통계량 살펴보기 
# describe() : 숫자 형태로 된 컬럼만 출력
mpg.describe()
```

```python
# 6. audi 데이터 접근 후 'hwy'컬럼 내림차순 정렬하여 5개 행만 보기 
# 불리언인덱싱, query(), 2차원.sort_valuews(by='컬럼',head()
audi = mpg[mpg['manufacturer'] == 'audi']
audi.sort_values(by = 'hwy',ascending = False).head() # 불리언 인덱싱으로 풀기
mpg.query('manufacturer == "audi"').sort_values('hwy', ascending = False).head() #query로 풀기
```

```python
# Q1
# mpg[['category','cty']].groupby('category').mean()
cty_avr = mpg.groupby('category').agg(cty_avr = ('cty', 'mean'))
```

```python
cty_avr.sort_values(by='cty_avr', ascending = False)
```

```python
# Q3
mf_hwy_avr = mpg.groupby('manufacturer').agg(mf_hwy_avr = ('hwy', 'mean'))
mf_hwy_avr.sort_values(by = 'mf_hwy_avr', ascending = False).head(3)
# mf_hwy_avr.sort_values(by = 'mf_hwy_avr', ascending = False).iloc[:3] (가능)
```

```python
# Q4 불리언 인덱싱
# compact 차종 정보 접근 > 회사별로 묶기 > 카운트 함수 > 결과 정렬

compact = mpg[mpg['category'] == 'compact']
result = compact.groupby('manufacturer').agg(cmp_cnt = ('category','count'))
result.sort_values('cmp_cnt', ascending = False)
```

```python
# Q4. 쿼리로 풀기 
# compact 차종 정보 접근 > 회사별로 묶기 > 카운트 함수 > 결과 정렬

mpg.query('category == "compact"')['manufacturer'].value_counts()
```

```python
# Q4. 내가 푼 것 

mf_compact = mpg.query('category == "compact"').groupby('manufacturer').agg(mf_compact = ('category', 'count'))
mf_compact.sort_values(by = 'mf_compact', ascending = False)
```
