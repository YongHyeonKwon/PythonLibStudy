```python
# 조건에 맞는 데이터 접근(불리언 인덱싱)
# 참True,거짓False
# 불리언 인덱싱 예시 데이터 생성
sample_arr=[True,False]
np.random.seed(1)
bool_arr = np.random.choice(sample_arr, size = 10)
display(bool_arr)

tmp_arr = np.arange(0,10)
display(tmp_arr)

# 불리언 인덱싱
display(tmp_arr[bool_arr])
```

```python
# 불리언 인덱싱
# 성적을 담은 배열, 학생 번호를 담은 배열을 생성해보자
# 50,100까지 담은 배열 생성하기, 1차원 배열로 요소값 20
np.random.seed(3)
big_score = np.random.randint(50,101,size=20)

stu_num = np.arange(1,21)
display(big_score)
display(stu_num)
```

```python
# 논리연산자를 활용해서 big_score 배열 중 80이상인 성적 확인하기
display(big_score[big_score >=80])

# 70미만인 성적은 몇개일까?
# 요소 개수 세어 주는 키워드 size
big_score[big_score<70].size
```

```python
# 성적이 60점 이상인 학생 번호 출력하기
# 성적이 60점 이상인가? 질문 True, False
# 학생 번호 배열에 인덱싱하기
stu_num[big_score>=60]
```

```python
data = np.loadtxt('./data/height_weight.txt',delimiter = ',')
data
```

```python
# 2) 정보 확인하기
# 배열 크기, 데이터 타입, 차원 크기

print(f'행{data.shape[0]}, 열:{data.shape[1]}')
print('데이터 타입: ', data.dtype)
print('차원 크기: ', data.ndim)
```

```python
data
```

```python
# 0 인덱스 번호로 접근하기 
bmi_arr = np.arange(1,11)

# 1. 키 변수에 키 데이터 10개 담기 
height = data[0]
# 2 몸무게 변수에 몸무게 데이터 10개 담기
weight = data[1]
display(height)
display(weight)

# 3. 키 데이터 단위 변경하기 cm -> m
# 1m = 100cm
# /100, *0.01
height2 = height / 100
display(height2)

# 4. 배열간의 연산을 통해서 bmi 공식 대입하여 결과 뽑아내기 
bmi_result = weight/(height2*height2) # or height**2
bmi_result

# 5. 과체중 이상인 데이터들 접근하기 (불리언 인덱싱)
# 몇 명일까? 6명
bmi_result[bmi_result>=23].size

# 6. 과체중만 접근하기
bmi_result[(bmi_result>=23) & (bmi_result<25)]

# 인덱스 번호로 뽑아내기
bmi_arr[(bmi_result>=23) & (bmi_result<25)]
```

### Numpy 유용한 함수

```python
np.random.seed(1)
arr = np.random.randint(1,11, size = (3,3))
arr

# sum(): 총합 함수
print(arr.sum()) # arr 배열 객체가 가지고 있는 메소드
print(np.sum(arr)) # np 자체 함수
```

```python
# mean(): 평균 함수
arr = np.array([6,9,6,1,1,])
print(arr.mean())
print(np.mean(arr))
```

```python
np.mean([1,2,3]) # np.mean 배열 객체가 아니어도 활용 가능
```

```python
# sqrt: 제곱근 함수
# square root : 제곱근 -> sqrt
sq_arr = np.array([4,16])
display(np.sqrt(sq_arr))
# display(sq_arr.sqrt()) -> 에러
```

```python
# abs(): 절대값 함수
num_arr = np.array([-1,2,3,-4])
np.abs(num_arr)
```

```python
# bincount(): 값을 카운팅하는 함수
np.random.seed(1)
rd_arr = np.random.randint(1,6, size = 5)
display(rd_arr)
# 각 데이터 빈도수 카운팅
np.bincount(rd_arr)
```

```python
# unique(): 중복없이 유니크한 값 확인하는 함수
np.unique(rd_arr, return_counts = True)
```

```python
rd_arr
```

```python
# where() : 조건에 해당하는 값 찾고 처리하는 함수
# 엑셀 if(조건,참반환,거짓반환)
# rd_arr 3이상인게 있는가? -> 인덱스 번호 반환
np.where(rd_arr>=3)
# 참이라면 1로 바꾸고, 거짓이라면 0이라고 바꾸자
np.where(rd_arr>=3,1,0)
```
