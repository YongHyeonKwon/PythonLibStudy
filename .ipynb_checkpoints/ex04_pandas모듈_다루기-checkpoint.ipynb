#### Pandas 연산 및 유용한 함수

```python
# 파일 불러오기 및 저장
# exam = pd.read_csv('경로 및 파일명', 
#                                    encoding = 'euc-kr',index_col = 'id')
# index_col : 해당 컬럼값을 인덱스값으로 활용하는 키워드
exam = pd.read_csv('./data/exam.csv')
exam
```

```python
# 데이터 파악하는 함수들
# 1. head(): 앞에서 5개 행 출력
exam.head()
# 2. tail(): 끝에서 5개 행 출력(5개 행 기본값, 변경 가능)
exam.tail(3)
# 3. info(): 데이터 속성 정보 출력
exam.info()
# shape: 행, 열 등의 배열 크기 확인
exam.shape
# 5. describe(): 요약 통계량 확인
exam.describe()
```

```python
# 데이터 정렬하기 
# 정렬 -> sort
# exam 데이터의 수학 컬럼 정렬해보기 
# ascending = False -> 내림차순 설정
exam['math'].sort_values(ascending = False)
# exam.loc[:,'math']
# exam.iloc[:.]
```

```python
exam.sort_values(by = 'math', ascending = False)
```

```python
# 총합을 구하는 함수 : sum()
# axis = 0 , 행방향으로 데이터 확인 == 열단위 처리 - sum 안에 씀
# axis = 1, 열방향으로 데이터 확인 == 행단위 처리 - sum 안에 씀
# 학생들의 수학, 영어, 과학 점수 총합 구하기 
exam[['math','english','science']]
exam.loc[:,'math':]
exam.iloc[:,2:].sum(axis = 1)
# 총합 컬럼 추가 : 파생변수(컬럼) 추가
exam['total'] = exam.iloc[:,2:].sum(axis = 1)
exam
```

```python
# 평균 구하는 함수 : mean()
# 각 학생의 3과목 평균 구하기 
exam['mean'] = exam.iloc[:,2:5].mean(axis = 1)
exam
```

```python
# 파생변수(컬럼) 추가2 - df.assign(컬럼명 = 새로운 데이터(만드는 공식))
# 실제 변수에 저장되는 것은 아님
# assign 함수 안에서 여러개의 컬럼을 생성할 수 있음
exam = exam.assign(mean2 = exam.iloc[:,2:5].mean(axis = 1))
exam              # 컬럼명        새 데이터 공식         ,컴마 사용해서 여러개 추가 가능.

# mean2 컬럼 삭제하기  drop 사용시 (대입도 가능 inplace = True도 가능)
exam.drop('mean2', axis = 1, inplace = True)
exam
```

```python
import numpy as np
# np.where(조건, 참, 거짓)
exam = exam.assign(result = np.where(exam['mean'] >= 60, 'pass', 'fail'))
exam
```

```python
# 데이터에 복잡한 기능 연결하여 처리하는 함수: apply()
# 1. def로 사용자가 정의하는 함수 생성
# 2. apply로 판다스 데이터에 연결
exam2 = exam.iloc[:,:5]
exam2
# 학생별 3과목의 평균을 구해서 85점 이상일 때 A, 70점 이상일 때 B, 60점 이상일 때 C, 나머지는 F
# 1. 예시 함수 만들기 
def grade_check(data):
    # axix = 1 했기 때문에 data 변수에는 한 학생의 데이터가 처리 단위로 연결됨
    mean = (data['math']+data['english']+data['science'])/3
    # 등급 판별 조건식
    if mean >= 85 :
        return 'A'
    elif mean >= 70 :
        return 'B'
    elif mean >= 60 :
        return 'C'
    else:
        return 'F'
    
exam2['grade'] = exam2.apply(grade_check, axis = 1)
exam2
```

```python
# 그룹별(집단별)로 요약 통계량 확인하는 함수 : groupby 
# groupby -> 집단별로 묶어줌 !
# 반별 수학 평균 구하기 
exam2[['nclass', 'math']].groupby('nclass').mean()
# 요약 통계 함수: mean(), sum(), median(), min(), max(), count()

# 반별 과학 중앙값 구하기
exam2[['nclass', 'science']].groupby('nclass').median()

# exam2
exam2.groupby('nclass')['science'].median()
```

```python
# groupby 활용해서 등급별 사람 수 확인하기 
# grade 집단별로 묶고 id컬럼에 count 통계 함수 연결
display(exam2[['grade', 'id']].groupby('grade').count())
exam2.groupby('grade')['id'].count()
```

```python
# 빈도수를 세주는 함수
# value_counts()
exam2['grade'].value_counts()
```

```python
# groupby랑 연결해서 사용되는 함수 agg
# 요약 통계를 구할 때 사용하는 함수 : agg()
# 요약 통계량 함수 : mean, sum, median, min, max, count, std
# 전체 수학점수 평균 구하기 
# exam2.agg(mean_math = (컬럼, 어떤 통계함수))
exam2.agg(mean_math = ('math', 'mean'))
# 반별 수학 평균 구하기 
exam2.groupby('nclass').agg(nclass_mean_math = ('math', 'mean'))
a = exam2.groupby('nclass').agg(nclass_mean_math = ('math', 'mean'))

# 위의 결과 내림차순 정렬해보기 
# 어떤 반이 수학 평균이 제일 높은가?
a.sort_values(by='nclass_mean_math', ascending = False)
```

```python
exam2.groupby('nclass').agg(nclass_mean_math = ('math', 'mean'),
                            nclass_mean_sc = ('science', 'mean'),
                            nclass_mean_eg = ('english', 'mean'))
```

```python
# 집단별(그룹별)로 묶고 다시 집단별로 묶기 
# 반별로 grade 빈도수 출력하기 
result = exam2.groupby(['nclass','grade']).agg(grade_cnt = ('grade','count'))
result
# 인덱스가 2개, 중복인덱스 (멀티인덱스)에 데이터 접근해보기 
# 4반의 C등급 사람의 수는?
result.loc[4,'C'].values[0]
result.loc[(4,'C'), 'grade_cnt']
```

```python
# 카테고리 형식으로 데이터 정리하기 
# 나이 데이터를 연령층으로 변경해보기 
ages = [0,2,15,21,23,37,51,61,20,48,100]
labels = ['미성년자', '청년', '중년', '장년', '노년'] # 구간에 대한 이름 설정 
bins = [-1,15,25,35,60,100] # 앞에는 미포함, 뒤에는 포함
# (-1,15], (15,25], (25,35] , ...
cats = pd.cut(ages, bins = bins, labels = labels)
cats

df_age = pd.DataFrame(ages, columns = ['ages'])
df_age['cats'] = cats # 파생변수 추가
```

```python
# 우리가 가진 나이 데이터는 카테고리별로 몇 개씩 있을까?
df_age['cats'].value_counts()
```

```python
# 병합하는 함수 : concat 
g1 = pd.DataFrame({'id':[1,2,3,4,5],'test':[60,80,70,90,85]})
g2 = pd.DataFrame({'id':[1,2,3,4,5],'test':[70,83,65,95,80]})
display(g1)
display(g2)
```

```python
result = pd.concat([g1,g2], axis = 1)
result
```

```python
# 컬럼명 변경하기 
result.columns= ['id1', 'test', 'id', 'test']
result
# 컬럼명 개별적으로 변경하기 
result = result.rename(columns = {'id1' : 'id2'})
result
```

```python
# id 컬럼 삭제 : drop()
# axis = 1 열방향으로 열의 이름을 확인 
result.drop('id', axis = 1, inplace = True)
result # 상태를 저장하려면 = , 함수 안에 inplace = True
```
