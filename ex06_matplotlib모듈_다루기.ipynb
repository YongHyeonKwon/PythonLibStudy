```python
# 데이터 불러와서 economi 변수에 담아보기 
# 위에서 행 5개 출력해보기
economi = pd.read_csv('./data/시도_성별_경제활동인구_총괄_KOSIS_2022.csv', 
                      encoding = 'euc-kr')
economi.head()
```

```python
# 데이터 전체 속성 정보 확인하기 
# economi.info()
# 행 34, 열 10
# 고용률, 실업률 컬럼 결측치 o
# 결측치 개수를 확인하는 방법 
economi.isnull().sum()
# 고용률 : 생산가능연령인 15세 이상인 인구 중에서 특정시점에 취업하고 있는 사람의 비율
# 실업률 : 경제활동참가자 중에서 실업상태에 있는 사람의 비율 
```

```python
# 고용률 결측치 처리해보기 
# 특성공학 : 기존의 컬럼 활용해서 데이터 생성
# 고용률 : 취업자 / 15세 이상인구 컬럼

# 15세 이상 인구 컬럼(열)접근 
economi['15세이상인구 (천명)'] 
# economi.loc[:,'15세이상인구 (천명)']
# economi.iloc[:, 2]

# 취업자 컬럼(열) 접근
economi['취업자 (천명)'] 

emp_pt = economi['취업자 (천명)'] / economi['15세이상인구 (천명)'] *100
economi['고용률(%)'] = emp_pt
economi.isnull().sum()
```

```python
# 실업률 처리해보기 
# 실업률 => 실업자/경제활동인구 
# economi 실업률 컬럼에 대입까지 ! 
economi['실업자 (천명)'] 
economi['경제활동인구 (천명)'] 
unemp_pt = economi['실업자 (천명)'] / economi['경제활동인구 (천명)'] * 100
economi['실업률(%)'] = unemp_pt
economi.isnull().sum()
```

```python
economi
```

```python
# 시도별 고용률 평균 구해보기 
# 시도별 고용률 평균이 가장 높은 지역 5개 확인해보기 
# 1. 필요한 컬럼들은?
# 2. groupby()
# 3. mean()
# 4. 상위 5개 데이터만 출력해보기 

economi[['행정구역(시도)', '고용률(%)']].groupby('행정구역(시도)').mean()
emp_mean = economi[['행정구역(시도)', '고용률(%)']].groupby('행정구역(시도)').agg(고용률평균 = ('고용률(%)','mean'))
emp_mean.sort_values(by = '고용률평균', ascending = False).head()
```

```python
x = emp_mean.index # 1차원
y = emp_mean['고용률평균'] # 1차원

plt.figure(figsize = (6,3))
plt.bar(x,y)
plt.xticks(rotation = 45) # x 축 이름 회전하기
plt.ylim(55,80)
plt.title('시도별 고용률 평균 시각화')
plt.xlabel('행정구역(시도)')
plt.ylabel('고용률 평균(%)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (8,4))
sns.barplot(x=x, y= y) # 색상 파레트 설정이 조금 더 다양함
plt.xticks(rotation = 45)
plt.show()
```

```python
#  남자의 데이터 접근 1. 불리언 인덱싱 2. query()
economi[economi['성별'] == '남자']['실업률(%)']
eco_m_unemp = economi.loc[economi['성별'] == '남자',['행정구역(시도)','실업률(%)']]
eco_m_unemp
# 2차원배열[불리언행,열]

x = eco_m_unemp['행정구역(시도)']
y = eco_m_unemp['실업률(%)']

sns.barplot(x=x, y=y)
plt.xticks(rotation = 45)
# plt.ylim(2.5,3.5)
plt.grid()
plt.show()
```

```python
# 우리가 푼 것.
a = economi[['행정구역(시도)', '성별', '실업률(%)']].groupby(['행정구역(시도)','성별']).mean()
a2=a.reset_index()
a3 = a2[a2.iloc[:,1]=='남자']

a4 = a3['실업률(%)']

a4

y2 = a4

plt.figure(figsize = (8,4))
sns.barplot(x=x, y= y2) # 색상 파레트 설정이 조금 더 다양함
plt.xticks(rotation = 45)
plt.show()
```

#### Scatter plot(산점도 그래프), Pie plot

```python
# 산점도 그래프 그려보기 
# 산점도 점 == 버블 (버블 사이즈, 버블 색상 설정)
x = np.arange(1,4) # 1,2,3
np.random.seed(2)
y = np.random.rand(3)# 실수 난수 생성
y
display(x)
display(y)

plt.scatter(x,y, s = [100,100,300], c = ['r','r','c'])
plt.show()
```
